## Inheritanse

```python
# inheritanse
class Area():
  def __init__(self, pin, location):
    self.pin = pin
    self.location = location
```

```python
# Now inheriting from Area class
class House(Area):
  
  def __init__(self,pin,location,color,rooms):
    self.color = color
    self.rooms = rooms
    Area.__init__(self,pin,location) # inheriting thse attributes from parent class itself. No need to define it here.
  def open_door(self):
    if self.color == "white" and self.rooms == 2:
      print("opening the door")
    else:
      print("not a white house. Could not open the door.")
```

```python
# Now inheriting from Area class
class House(Area):
  
  def __init__(self,pin,location,color,rooms):
    self.color = color
    self.rooms = rooms
    super().__init__(pin,location) # inheriting thse attributes from parent class itself. No need to define it here.
  def open_door(self):
    if self.color == "white" and self.rooms == 2:
      print("opening the door")
    else:
      print("not a white house. Could not open the door.")
```

```python
small_house = House(color='white',rooms = 2,pin=700016,location="Kolkata")
```

```python
small_house.open_door()
```

```python
print(small_house.color, small_house.rooms,small_house.pin,small_house.location)
```

```python
# Now inheriting from Area class
class House(Area):
  area = "downtown" #class object attribute
  def __init__(self,pin,location,color,rooms):
    self.color = color
    self.rooms = rooms
    Area.__init__(self,pin,location) # inheriting thse attributes from parent class itself. No need to define it here.
  def open_door(self, key): # adding a argument relative to this method only
    if self.color == "white" and self.rooms == 2 and key=="yes":
      print("opening the door") # you do not need to add self for the key. 
    else:
      print("not a white house. Could not open the door.")
```

```python
white_house = House(color='white',rooms = 2,pin=700016,location="Kolkata")
```

```python
white_house.area
```

```python
white_house.open_door(key='yes')
```

## Polymorphism

```python
# Polymorphism
# when two different classes have same method name, we can do the polimorphysm in it
class House():
  def __init__(self, color):
    self.color = color
  def house_color(self):
    print(f"House color is {self.color}")

class CommunityHouse():
  def __init__(self, community_house_color):
    self.community_house_color = community_house_color
  def house_color(self):
    print(f"House color is {self.community_house_color}")
```

```python
black_house = House(color="Black")
White_house = CommunityHouse(community_house_color="White")
```

```python
def give_house_color(house_name):
  return house_name.house_color()
```

```python
give_house_color(black_house)
```
